```python
train = pd.read_csv("data/train.csv")
test = pd.read_csv("data/test.csv")
```

```python
def parse_date(data,date_name='Dates',year_name='Year', month_name='Month', week_name='Week',\
               minute_name='Minute',hour_name='Hour'):
    
    data[year_name] = data[date_name].dt.year
    data[month_name] = data[date_name].dt.month
    data[week_name] = data[date_name].dt.week
    data[hour_name] = data[date_name].dt.hour
    data[minute_name] = data[date_name].dt.minute
    return data
```

```python
train['Dates'] = pd.to_datetime(train['Dates'])
test['Dates'] = pd.to_datetime(test['Dates'])
train = parse_date(train)
test = parse_date(test)
```

```python
def generate_count_and_distribution(data, base_name, by):
    f1 = base_name + '_count'
    f2 = base_name + '_distribution'
    data[f1] = 0
    
    if type(by) is str:
        by = [by]
   
    target_data = data[by+[f1]+['Category']].groupby(by=(by+['Category'])).agg({f1:'count'})
    if len(by)!=0:
        data[f2] = 0
        temp_data = data[by+[f2]].groupby(by=by).agg({f2:'count'})
        target_data[f2] = target_data[f1]/temp_data[f2]
    target_data.reset_index(inplace=True) 
    return target_data
```
